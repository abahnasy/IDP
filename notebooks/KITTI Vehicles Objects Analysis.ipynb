### Analysis: Vehicles in KITTI Dataset
Analysis is based on the training dataset

```python
import numpy as np
import pickle
import matplotlib.pyplot as plt
```

```python
# Move up one directory
file = open('../kitti_vechicle_num_pts.obj', 'rb')
pts_list = pickle.load(file)
```

```python
print('Total number of Vehicles in the dataset: {}'.format(len(pts_list)))
```

```python
# Statistics for the number of points in the bounding box
a = np.array(pts_list)
print('max points per bbox: {}, Min Points per bbox: {}, Average Points per bbox: {}'.format(a.max(), a.min(), int(a.mean())))
```

```python
plt.figure(figsize=(20,10))
plt.title('No of points inside Bounding box')
plt.xlabel('No of points')
plt.ylabel('Frequency')
plt.grid()
h = plt.hist(a, bins=[i for i in range(0,1000,10)])
```

```python
fig, axs = plt.subplots(5, 2, figsize=(15,30))
plt.subplots_adjust(top=0.95, hspace=0.3, wspace=0.2)
fig.suptitle('ff')
ranges = [1, 50, 100, 500, 1000, 2000, 3000, 4000, 5000, 6000, 7000]
steps = [5, 5, 10, 10, 100, 100, 100, 100, 100, 100]
assert len(steps)+1 == len(ranges)

for idx in range(len(steps)):
    i = idx//2
    j = idx % 2
#     print('processing plot ({}, {})'.format(i, j))
    axs[i, j].set_title('Freq of bboxes with points in range {} to {}, total bboxes are {}'.format(ranges[idx], ranges[idx+1], np.logical_and(a >= ranges[idx], a < ranges[idx+1]).sum()))
    axs[i, j].grid()
    axs[i, j].hist(a, bins=[i for i in range(ranges[idx], ranges[idx+1], steps[idx])])
# for ax in axs.flat:
#     ax.set(xlabel='No of Points in bbox', ylabel='Frequency')
# Hide x labels and tick labels for top plots and y ticks for right plots.
# for ax in axs.flat:
#     ax.label_outer()
```
